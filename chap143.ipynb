#### Geometric Transformations of Images

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('opcv.png',0)
rows,cols = img.shape
res1 = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)
res2 = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)
res3 = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)
```

```python
cv2.namedWindow('res1')
cv2.namedWindow('res2')
cv2.namedWindow('res3')
res = cv2.resizeWindow('res1',rows*2, cols*2)
res2 = cv2.resizeWindow('res2',rows*2, cols*2)
res3 = cv2.resizeWindow('res3',rows*2, cols*2)

while(1):
    cv2.imshow('res1',res1)
    cv2.imshow('res2',res2)
    cv2.imshow('res3',res3)
    if cv2.waitKey(20) & 0xFF == 27:
        break
cv2.destroyAllWindows()
```

```python
img = cv2.imread('opcv.png',0)
rows,cols = img.shape
M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))
cv2.imshow('img',img)
cv2.imshow('det',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
M = cv2.getRotationMatrix2D((cols/2,rows/2),45,1)
dst = cv2.warpAffine(img,M,(cols,rows))
cv2.imshow('img',img)
cv2.imshow('det',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
##### Perspective Transformation
```

```python
img = cv2.imread('dsu1.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[323,282],[760,310],[765,509],[322,534]])
pts2 = np.float32([[300,300],[800,300],[800,600],[300,600]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(1100,900))
```

```python
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
